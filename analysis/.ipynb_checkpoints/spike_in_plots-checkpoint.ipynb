# Spike in plots 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spike_in_data = pd.read_csv('spikein_data.csv')
# spike_in_data = spike_in_data[spike_in_data.SpikeinID=='SDSI_01']
```

```python
spike_in_data = pd.read_csv('to_plot.csv')
print(spike_in_data)
```

```python
amplicon_coverage = pd.read_csv('amplicon_coverage.csv')
```

```python
amplicon_coverage.head()
```

```python
spike_in_data.head()
```

```python
# Filter amplicon data to only include 1A and 5, this also drops species  
with open('pool_5.txt') as f:
    pool_5_loci = f.read().splitlines()
species_loci = ['PmUG01_12_v1-1397996-1398245-1AB','PocGH01_12_v1-1106456-1106697-1AB','PvP01_12_v1-1184983-1185208-1AB']
amplicon_coverage = amplicon_coverage[(amplicon_coverage.Locus.str.endswith('-1A')) | (amplicon_coverage.Locus.isin(pool_5_loci))]
amplicon_coverage = amplicon_coverage[~amplicon_coverage.Locus.isin(species_loci)]
```

```python
# Average Reads per sampleID per locus 
avg_coverage_per_sample_per_locus = amplicon_coverage.groupby(['SampleID','Locus']).Reads.mean().reset_index()
```

```python
# Plot average cov per sample 

# Get averages across replicate samples
avg_coverage_per_sample_per_locus = amplicon_coverage.groupby(['SampleID','Locus']).Reads.mean().reset_index()
avg_spikein_reads = spike_in_data.groupby('SampleID')['Reads'].mean().reset_index()

# Check unique sample IDs
avg_coverage_sample_ids = set(avg_coverage_per_sample_per_locus['SampleID'])
spike_in_sample_ids = set(avg_spikein_reads['SampleID'])
if avg_coverage_sample_ids != spike_in_sample_ids:
    print("Sample IDs do not match between datasets.")

# Sort DataFrames by SampleID
avg_coverage_per_sample_per_locus.sort_values(by='SampleID', inplace=True)
avg_spikein_reads.sort_values(by='SampleID', inplace=True)

# Group reads by SampleID to box plot
grouped_reads = avg_coverage_per_sample_per_locus.groupby('SampleID')['Reads']

# Create a box plot for each sample
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
boxplot = plt.boxplot([group for _, group in grouped_reads], labels=[name for name, _ in grouped_reads])
# Plot the spike in reads
x_coordinates = [boxplot['boxes'][i].get_xdata().mean() for i in range(len(avg_coverage_sample_ids))]
plt.scatter(x_coordinates, avg_spikein_reads['Reads'], color='red', label='Spike-in Reads')

plt.xlabel('Sample ID')
plt.ylabel('Reads')
# plt.legend()
plt.yscale('log')
plt.title('Reads per Experiment for Amplicons and Spike-ins')
plt.xticks(rotation=90)
plt.savefig('spike_in_and_amplicon_reads.png')
plt.show()
```
